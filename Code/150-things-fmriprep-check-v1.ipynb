```python
import os
import pandas as pd
import numpy as np
import ants
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
indir = '../Data/things/fmriprep2/'
subs = ['sub-01','sub-02','sub-03']
sub = 'sub-01'
```

```python
#BC-ABCD-denoise/Data/things/fmriprep2/sub-01/ses-things01/func/
```

```python
def is_path(sub,r,sess):
    #sess = 2
    #r = 2
    session = f'things{sess:02d}'
    fn = f'{sub}_ses-{session}_task-things_run-{r:02d}_space-MNI152NLin2009cAsym_res-2_desc-preproc_bold.nii.gz'
    #fn = f'sub-01_ses-localizer2_task-pRF_run-01_space-MNI152NLin2009cAsym_res-2_desc-preproc_bold.nii.gz'
    path = os.path.join(indir,sub,'ses-'+session,'func',fn)
    return os.path.exists(path)
```

```python
sub = 'sub-01'
rvec = np.arange(1,12)
sess_vec = np.arange(1,14)
is_done = np.array([[is_path(sub,r,sess) for r in rvec] for sess in sess_vec])
plt.figure(figsize=(15,5))
sns.heatmap(is_done,cbar=False,annot=True,vmin=0,vmax=1);
plt.yticks(ticks=np.arange(len(sess_vec))+.5,labels=sess_vec);plt.ylabel('session')
plt.xticks(ticks=np.arange(len(rvec))+.5,labels=rvec);plt.xlabel('run')
plt.title(sub,fontsize=16)
```
