```python
import os
import numpy as np
import pandas as pd
```

```python
indir = '../Data/020-fmriprepped/'
subs_orig = [s for s in os.listdir(indir) if all((s.startswith('sub'),not s.endswith('.html')))]
n_orig = len(subs_orig)
epi_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/func/{sub}_ses-baselineYear1Arm1_task-nback_run-0{r}_space-MNI152NLin2009cAsym_res-2_desc-preproc_bold.nii')
cf_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/anat/analysis_mask_CF.nii')
gm_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/anat/analysis_mask_GM.nii')
compcorr_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/func/{sub}_ses-baselineYear1Arm1_task-nback_run-01_space-MNI152NLin2009cAsym_res-2_desc-COMPCORR_bold.nii')
deepcorr_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/func/{sub}_ses-baselineYear1Arm1_task-nback_run-01_space-MNI152NLin2009cAsym_res-2_desc-DEEPCOR_bold.nii')

s8_epi_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/func/s8_{sub}_ses-baselineYear1Arm1_task-nback_run-0{r}_space-MNI152NLin2009cAsym_res-2_desc-preproc_bold.nii')
s8_compcorr_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/func/s8_{sub}_ses-baselineYear1Arm1_task-nback_run-0{r}_space-MNI152NLin2009cAsym_res-2_desc-COMPCORR_bold.nii')
s8_deepcorr_fn = os.path.join(indir,'{sub}/ses-baselineYear1Arm1/func/s8_{sub}_ses-baselineYear1Arm1_task-nback_run-0{r}_space-MNI152NLin2009cAsym_res-2_desc-DEEPCOR_bold.nii')

subs = [sub for sub in subs_orig if all(( 
    
    os.path.exists(cf_fn.format(sub=sub)),
    os.path.exists(gm_fn.format(sub=sub)),

    os.path.exists(epi_fn.format(sub=sub,r=1)),
    os.path.exists(epi_fn.format(sub=sub,r=2)),
    os.path.exists(s8_epi_fn.format(sub=sub,r=1)),
    os.path.exists(s8_epi_fn.format(sub=sub,r=2)),                                        
    os.path.exists(compcorr_fn.format(sub=sub,r=1)),
    os.path.exists(compcorr_fn.format(sub=sub,r=2)),
    os.path.exists(s8_compcorr_fn.format(sub=sub,r=1)),
    os.path.exists(s8_compcorr_fn.format(sub=sub,r=2)),
    os.path.exists(deepcorr_fn.format(sub=sub,r=1)),
    os.path.exists(deepcorr_fn.format(sub=sub,r=2)),
    os.path.exists(s8_deepcorr_fn.format(sub=sub,r=1)),
    os.path.exists(s8_deepcorr_fn.format(sub=sub,r=2)),
        
    ))]

subs.sort()
print(f'{len(subs)}/{len(subs_orig)}')
```

```python
check = []
```

```python
for sub in subs_orig:
    val = [
        os.path.exists(cf_fn.format(sub=sub)),
        os.path.exists(gm_fn.format(sub=sub)),
        os.path.exists(epi_fn.format(sub=sub,r=1)),
        os.path.exists(epi_fn.format(sub=sub,r=2)),
                                                
        os.path.exists(compcorr_fn.format(sub=sub,r=1)),
        os.path.exists(compcorr_fn.format(sub=sub,r=2)),
        os.path.exists(deepcorr_fn.format(sub=sub,r=1)),
        os.path.exists(deepcorr_fn.format(sub=sub,r=2)),
        
        os.path.exists(s8_epi_fn.format(sub=sub,r=1)),
        os.path.exists(s8_epi_fn.format(sub=sub,r=2)),
        os.path.exists(s8_compcorr_fn.format(sub=sub,r=1)),
        os.path.exists(s8_compcorr_fn.format(sub=sub,r=2)),
        
        os.path.exists(s8_deepcorr_fn.format(sub=sub,r=1)),
        os.path.exists(s8_deepcorr_fn.format(sub=sub,r=2))]

    check.append( val )
```

```python
colnames = ['cf_fn',
'gm_fn',
'epi_fn',
'epi_fn',
'compcorr_fn',
'compcorr_fn',
'deepcorr_fn',
'deepcorr_fn',
's8_epi_fn',
's8_epi_fn',
's8_compcorr_fn',
's8_compcorr_fn',
's8_deepcorr_fn',
's8_deepcorr_fn',]
```
