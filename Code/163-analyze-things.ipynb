```python
import os
import numpy as np
import pandas as pd
import ants
```

```python
ofdir_root = '../Data/StudyForrest/ensembles_last_CVAE'
analysis_name = "refactored-THINGS-exp-v1"
ofdir = os.path.join(ofdir_root,analysis_name)
```

```python
all_files = os.listdir(ofdir)
```

```python
#im_name = 'corr2living'
for im_name in ['corr2living','corr2nonliving','contrast_living','contrast_nonliving']:
    for kind in ['preproc','compcor','deepcor']:
        files = [os.path.join(ofdir,file) for file in all_files if all((file.startswith('corr2living'),file.endswith(f'_{kind}.nii.gz')))]
        im = ants.image_read(files[0])
        arr = np.array([ants.image_read(file).numpy() for file in files]).mean(axis=0)
        im.new_image_like(arr).to_filename(os.path.join(ofdir,f'{im_name}_{kind}_avg.nii.gz'))
```
