```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
import shutil
import tarfile
from tqdm import tqdm
```

```python
#indir = '../Data/000-Raw/fmriresults01/abcd-mproc-release5/'
indir = '../Data/ABCDfMRInbackall/fmriresults01/abcd-mproc-release5/'
fns = [f for f in os.listdir(indir) if not f.endswith('.tgz.partial')]

fns[0:5]
```

```python
anat_fn_temp = '../../BC-ABCD-ADHD/Data/ABCD-anat/sub-{sub}/ses-baselineYear1Arm1/anat/sub-{sub}_ses-baselineYear1Arm1_run-01_T1w.nii'
```

```python
df = pd.DataFrame(np.array([val.split('_') for val in fns]),columns=['subID','eventName','taskID','filename'])
#df = df.iloc[[not val.endswith('.tgz.partial') for val in df['filename'].values]]
df['fn_func'] = fns
df = df.iloc[df['eventName'].values=='baselineYear1Arm1']
nscans = [sum(df['subID'].values==val) for val in df['subID'].values]
df['nscans'] = nscans
df = df.iloc[df['nscans'].values>=2]
has_anat = [os.path.exists(anat_fn_temp.format(sub=val)) for val in df['subID'].values]
anat_fn = [anat_fn_temp.format(sub=val) for val in df['subID'].values]
df['has_anat'] = has_anat
df['anat_fn'] = anat_fn
df = df.iloc[df['has_anat'].values]
df = df.sort_values(by='subID')
print(df.shape)
df
```

```python
'{} unique subjects'.format(len(pd.unique(df['subID'])))
```

```python
for i in tqdm(range(len(df))):
    func_fn = os.path.join(indir,df['fn_func'].values[i])
    assert os.path.exists(func_fn)
    tar = tarfile.open(func_fn, 'r:gz')
    for item in tar:
        tar.extract(item,'../Data/010-Data_bids2')
```

```python
unique_subs = pd.unique(df['subID'].values)
len(unique_subs)
```

```python
for sub in unique_subs:
    if not os.path.exists('../Data/010-Data_bids2/sub-{sub}/ses-baselineYear1Arm1/anat'.format(sub=sub)):
        shutil.copytree(src='../../BC-ABCD-ADHD/Data/ABCD-anat/sub-{sub}/ses-baselineYear1Arm1/anat/'.format(sub=sub),
                   dst='../Data/010-Data_bids2/sub-{sub}/ses-baselineYear1Arm1/anat'.format(sub=sub))
```

```python
for sub in unique_subs:
    temp1 = '../Data/010-Data_bids2/sub-{sub}/ses-baselineYear1Arm1/func/sub-{sub}_ses-baselineYear1Arm1_task-nback_run-01_motion.tsv'.format(sub=sub)
    if os.path.exists(temp1):
        shutil.copyfile(src=temp1,dst=os.path.join('../Data/015-subject-motion-files',temp1.split('/')[-1]))
        os.remove(temp1)
    
    temp2 = '../Data/010-Data_bids2/sub-{sub}/ses-baselineYear1Arm1/func/sub-{sub}_ses-baselineYear1Arm1_task-nback_run-02_motion.tsv'.format(sub=sub)
    if os.path.exists(temp2):
        shutil.copyfile(src=temp2,dst=os.path.join('../Data/015-subject-motion-files',temp2.split('/')[-1]))
        os.remove(temp2)
```
