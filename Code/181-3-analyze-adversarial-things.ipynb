```python
import os
import pandas as pd
import numpy as np
import ants
from matplotlib import pyplot as plt
```

```python
def get_roi_data(im_fn,roi_fn):
    roi = ants.image_read(roi_fn)
    im = ants.image_read(im_fn)

    roi_data = im.numpy()[roi.numpy()==1].mean()
    return roi_data
```

```python
indir = '../Data/StudyForrest/fmriprep/'
subs = [s for s in os.listdir(indir) if all((s.startswith('sub'),not s.endswith('.html')))]
subs.sort()
print(len(subs))
```

```python
ofdir_root = '../Data/StudyForrest/ensembles_last_CVAE'
analysis_name = 'refac-adversarial-forrest-v2'
analysis_name = 'neurips-adversarial-forrest-v1-beta-1e-5'
ofdir = os.path.join(ofdir_root,analysis_name)
```

```python
rvec=[1,2,3,4]
```

```python
anames = ['conv_denoise','conv_denoise_long','conv_denoise_large','conv_denoise_smooth', 'conv_denoise_unet']
anames = ['conv_denoise']

roi_fn_temp = '../Data/StudyForrest/ROIs/rFFA_final_mask_{sub}_bin.nii.gz'
res = np.zeros((len(rvec),len(subs),len(anames)))

for s in range(len(subs)):
    for r in rvec:
        for a in range(len(anames)):
            #kind = kinds[k]
            im_fn = os.path.join(ofdir,f'contrast_face_{s}_R{r}_{anames[a]}_deepcor.nii.gz')
            roi_fn = roi_fn_temp.format(sub=subs[s])
            res[r-1,s,a] = get_roi_data(im_fn,roi_fn)
temp = res.mean(axis=0).mean(axis=0)
for i,aname in enumerate(anames):
    print(f'{aname}: {temp[i]:.2f}')            
```
