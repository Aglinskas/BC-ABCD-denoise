```python
import os
import numpy as np
import pandas as pd
import ants
from tqdm import tqdm
```

```python
indir = '../Data/StudyForrest/ensembles_last_CVAE/refactored-THINGS-exp-v1/'
```

```python
'../Data/StudyForrest/ensembles_last_CVAE/refactored-THINGS-exp-v1/'
```

```python
name='corr2nonliving'
name = 'contrast_nonliving'
name = 'contrast_living'
for kind in ['preproc','compcor','deepcor']:

    files = [os.path.join(indir,f) for f in os.listdir(indir) if all((f.startswith(name),f.endswith(f'_{kind}.nii.gz')))]
    n = len(files)
    
    im = ants.image_read(files[0])
    arr = np.zeros(im.numpy().shape)
    for i in tqdm(range(n)):
        arr+=ants.image_read(files[i]).numpy()
    arr = arr/n
    im_avg = im.new_image_like(arr)
    im_avg.to_filename(f'../{name}_avg_N_{n}_{kind}.nii')
```
